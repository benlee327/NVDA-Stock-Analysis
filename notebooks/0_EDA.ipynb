```python
import pandas as pd
import matplotlib.pyplot as plt

# EDA(exploratory data analysis to check on data cleanliness)
df = pd.read_csv(
    "../data/yfinance_nvda.csv", 
    skiprows=3, #skipping filler label rows
    names=["Date","Close","High","Low","Open","Volume"], # Redefining column names for future use (prices are adjusted)
    parse_dates=True,
    index_col=0)
print(df.index.name)   # "Date"
print(df.columns)      # ["Close","High","Low","Open","Volume"]
df.head() # Show small sample data from csv
```

```python
df["Close"].describe()
```

```python
#Simple plot graph for close column
plt.figure(figsize=(10, 4))
plt.plot(df["Close"], label="NVIDIA Close")
plt.title("NVDA Adjusted Closing Price (2019â€“2025)")
plt.legend()
plt.show()
```
